# Analysing the correlation between feature and target

```python
plt.figure(figsize=(8,4))
sb.heatmap(iris.corr(),annot=True,cmap='Blues');
```

```python
X=iris[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
Y=iris[["Species"]]
```

# Spliting the datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
iris.columns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier(max_depth = 5, random_state = 0)
# Train the model on the data
clf.fit(X_train, Y_train)
# Predict labels of unseen (test) data
# clf.predict(X_test)
```

# Predictions

```python
predictions = clf.predict(X_test)
print("The accuracy of Decision Tree is:",metrics.accuracy_score(predictions,Y_test))
```

# Visualizing the Tree

```python
from sklearn import tree
tree.plot_tree(clf)
```
